```python
def loadData(self,folder_name='plain_road_potholes'):
    path = '../data/'+folder_name+'/'
    print(path)
    os.chdir(path)
    l_ = os.listdir()
    print(l_)
    df = pd.DataFrame(columns=['Time','Gx','Gy','Gz','Ax','Ay','Az'])
    for i in l_:
        temp = pd.read_csv(i)
        # print(temp)
        df = pd.concat([df,temp],axis=0)

    os.chdir('../../LoadData/')
    return df

loadData(0,'plain_road_unmarked_sb')
```
